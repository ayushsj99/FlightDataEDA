```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dftrain = pd.read_excel('Dataset/Data_Train.xlsx')
dftrain.head()
```

```python
dftest = pd.read_excel('Dataset/Test_set.xlsx')
dftest.head()
```

```python
dftrain.info()
dftest.info()
```

```python
df = pd.concat([dftrain,dftest], axis=0, ignore_index=True)
df.info()
```

```python
df.head()
```

```python
df.info()
```

```python
# Feature Engineering process
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
 df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x: x.split(':')[0]) 
df['Arrival_Minutes'] = df['Arrival_Time'].apply(lambda x: x.split(':')[1]) 
```

```python
df.head()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int) 
```

```python
df.info()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x: x.split(':')[0]) 
df['Dep_Minutes'] = df['Dep_Time'].apply(lambda x: x.split(':')[1]) 
```

```python
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int) 
```

```python
df.info()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1}) 
```

```python
df.info()
```

```python
df['Duration_in_mins'] = df['Duration'].apply(lambda x: 
    (int(x.split('h')[0]) * 60 if 'h' in x else 0) + 
    (int(x.split('h')[1].replace('m', '').strip()) if 'h' in x and 'm' in x else 
     int(x.replace('m', '').strip()) if 'm' in x else 0)
)
```
